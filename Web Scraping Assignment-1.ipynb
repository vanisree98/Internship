## 2.Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)and make data frame.


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
#requests page source from url
url = 'https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating'
page =requests.get(url)
page
soup=BeautifulSoup(page.content,'html.parser')
```

```python
movie_title=[]
year=[]
rating=[]
```

```python
#Scraping the movie_title
movie=soup.find_all('div',attrs={'class':'lister-item mode-advanced'})
for store in movie:
    name=store.h3.a.text
    movie_title.append(name)
#Scraping the year_of_release
    year_of_release=store.h3.find('span',class_="lister-item-year text-muted unbold").text.replace('(','').replace(')','')
    year.append(year_of_release)
#Scraping the rating    
    rate=store.find('div',class_='inline-block ratings-imdb-rating').text.replace('\n',' ')
    rating.append(rate)
```

```python
movie_df=pd.DataFrame({'movie_title': movie_title,'Rating':rating,'year of release':year})
movie_df
```

## 3) Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release) and make data frame.

```python
import re
url = 'https://www.imdb.com/india/top-rated-indian-movies/'
response = requests.get(url)
response
soup = BeautifulSoup(response.text, 'lxml')
movies = soup.select('td.titleColumn')
ratings = [b.attrs.get('data-value') for b in soup.select('td.posterColumn span[name=ir]')]
imdb_ind = []

for index in range(0, 100):
    
    movie_string = movies[index].get_text()
    movie = (' '.join(movie_string.split()).replace('.', ''))
    movie_title = movie[len(str(index))+1:-7]
    year = re.search('\((.*?)\)', movie_string).group(1)
    
    data = {"movie_title": movie_title,
            "year": year,
            "rating": ratings[index],
            }
    imdb_ind.append(data)

for item in imdb_ind:
  print(item['movie_title'], item['rating'], '('+item['year']+') -')
```

```python
import pandas as pd
df2 = pd.DataFrame(imdb_ind)
df2
```

```python
df2['rating']=df2['rating'].astype(float)
df2['rating']=df2['rating'].round(decimals=2)
df2.head()
```

## 7) Write a python program to scrape details of all the posts from http://coreyms.com/. Scrape the heading, date, content  and the code for the video from the link for the youtube video from the post.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

source = requests.get('http://coreyms.com').text

soup = BeautifulSoup(source,'html.parser')
```

```python
Headline=[]
Content=[]
DateS=[]
YT_link=[]

for article in soup.find_all('article'):
#Scraping the  Headline
    headline = article.h2.a.text
    Headline.append(headline)
    
#Scraping the content
    content = article.find('div', class_='entry-content').p.text
    Content.append(content)
#scraping the Date    
    Date= article.find('header', class_='entry-header').p.text[0:17]
    DateS.append(Date)
#scraping the youtube link   
    try:
        vid_src = article.find('iframe', class_='youtube-player')['src']

        vid_id = vid_src.split('/')[4]
        vid_id = vid_id.split('?')[0]

        yt_link = f'https://youtube.com/watch?v={vid_id}'
    except Exception as e:
        yt_link = None

    YT_link.append(yt_link)
```

```python
coreyMS=pd.DataFrame({})

coreyMS['headline']=Headline[0:11]
coreyMS['content']= Content[0:11]
coreyMS['Date']=DateS[0:11]
coreyMS['yt_link']=YT_link[0:11]

coreyMS
```
